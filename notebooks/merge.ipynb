```python
import pandas as pd 
```

```python
performance_df = pd.read_csv('../datasets/performance_data.csv')
category_df = pd.read_csv('../datasets/labeled.csv')
```

```python
category_df['game_id'] = category_df['game_id'].astype('object')
performance_df['game_id'] = performance_df['game_id'].astype('object')
```

```python
category_performance = performance_df.merge(category_df, on='game_id', how='inner')
```
